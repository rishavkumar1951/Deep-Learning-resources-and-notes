```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('3.concertriccir2.csv')
df.shape
```

```python
df.head()
```

```python
plt.scatter(df['x'],df['y'],c=df['class'])
```

```python
X = df.iloc[:,0:2].values
y = df.iloc[:,-1].values
```

# model building


```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
from tensorflow.keras.layers import BatchNormalization
```

```python
model = Sequential()

model.add(Dense(2,activation='relu',input_dim=2))
model.add(Dense(2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history1 = model.fit(X,y,epochs=200,validation_split=0.2)
```

```python
model = Sequential()

model.add(Dense(3,activation='relu',input_dim=2))
model.add(BatchNormalization())
model.add(Dense(2,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history2 = model.fit(X,y,epochs=200,validation_split=0.2)
```

```python
plt.plot(history1.history['val_accuracy'],color='black')  # without batch norm
plt.plot(history2.history['val_accuracy'],color='green')  # with Batch Norm
```

with batch normalization accuracy increasing in less epochs and major difference from not using batch Normalization
