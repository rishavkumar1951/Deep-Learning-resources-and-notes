```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/DailyDelhiClimateTrain.csv')
df.shape
```

```python
df.head()
```

```python
df = df[['date' ,'meantemp']].iloc[:114]
df.head()
```

```python
plt.scatter(df['date'] , df['meantemp'] , c = 'red')
```

```python
for i in [0.1 , 0.4 , 0.8 , 0.9, 1]:
  plt.figure (figsize = (15,6))
  df['meantemp'].ewm(alpha = i).mean().plot()
  plt.scatter(df['date'] , df['meantemp'] , c = 'red')
  plt.title(f"with alpha is {i} and beta = {1 - i}")
  plt.show()
```
