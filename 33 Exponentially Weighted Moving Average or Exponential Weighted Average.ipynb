```python
import pandas as pd 
import numpy as np 

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./3. Increasing model accuracy/DailyDelhiClimateTrain.csv')
```

```python
df =df[['date', 'meantemp']]
df
```

```python
plt.scatter(df['date'], df['meantemp'], color='red')
```

```python
x1 = df['meantemp'].ewm(alpha=0.9).mean()
```

```python
x1   # this gives EWMA for every row
```

```python
df['ewma'] = x1
```

```python
plt.scatter(df['date'], df['meantemp'],color='red')
plt.plot(df['date'],x1,color='black', label='alpha = 0.9')
plt.legend()
plt.show()
```
